# Spotify Explorer

```python
import pandas as pd
import altair as alt
import numpy as np
alt.data_transformers.disable_max_rows()
```

```python
data = pd.read_csv("../data/raw/spotify.csv")
```

## Specific artists popularity histogram

```python
data["track_artist"].unique()[0:5]
```

```python
click = alt.selection_multi(fields=['playlist_genre'], bind='legend')

chart = alt.Chart(data[data["track_artist"]=="Ed Sheeran"]).mark_bar().encode(
    x=alt.X('track_popularity', bin=True, title="Track popularity"),
    y=alt.Y('count()'),
    color="playlist_genre",
    opacity=alt.condition(click, alt.value(0.9), alt.value(0.2)),
    tooltip=alt.Tooltip('count()')
)

chart.add_selection(click)
```

```python
chart = alt.Chart(data[data["track_artist"]=="Ed Sheeran"]).mark_bar().encode(
    x=alt.X('track_popularity', bin=True, title="Track popularity"),
    y=alt.Y('count()'),
    tooltip=alt.Tooltip('count()')
)
rule = alt.Chart(data[data["track_artist"]=="Ed Sheeran"]).mark_rule(color='red').encode(
    x='mean(track_popularity):Q'
)

chart+rule
```
